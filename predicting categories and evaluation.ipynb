```python
from sklearn.linear_model import LogisticRegression

#Features "free" word and "!"
x = [
    [0, 0], [1, 0], [0, 1],   # Safe emails (Low counts)
    [5, 3], [10, 5], [3, 8],  # Spam emails (High counts)
    [1, 0], [0, 2],           # More Safe
    [6, 4], [8, 9]            # More Spam
]

# The target (1) spam and (0) safe

y = [0, 0, 0, 1, 1, 1, 0, 0, 1, 1]


# Instantiate (LogReg)

model = LogisticRegression()

# Fit
model.fit(x, y)

# New Email arrived 4, 5
new_email = [[2, 4]]

#prediction
prediction = model.predict(new_email)

#probability
probability = model.predict_proba(new_email)

print(f"The prediction:  {prediction[0]}            -- (o) safe and (1) spam")
print(f"The probability: {probability[0][1]*100: .0f} %")
```

```python
from sklearn.metrics import confusion_matrix

# Let's create a hypothetical test set
# Ytrue, lets put some actual values (0=safe and 1=spam)

y_true = [0, 1, 0, 1, 0, 1, 0, 0, 1, 1]

# Ypred is what our model predicted
y_pred = [0, 1, 0, 0, 0, 1, 1, 0, 1, 1]

# Generate the matrix
cm = confusion_matrix(y_true, y_pred)

print("Confusion Matrix")
print(cm)
```

```python
from sklearn.metrics import classification_report

# Using the same dummy data from before
y_true = [0, 1, 0, 1, 0, 1, 0, 0, 1, 1]
y_pred = [0, 1, 0, 0, 0, 1, 1, 0, 1, 1]

print(classification_report(y_true, y_pred, target_names=['safe','spam']))
```
